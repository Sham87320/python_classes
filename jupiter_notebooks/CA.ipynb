```python
#Task 2 (a)
import numpy as np
even_numbers_array = np.empty((4, 50, 5))
start_value = 2000
end_value = 4000
num_elements = (end_value - start_value) // 2  
even_numbers = np.arange(start_value, end_value, 2) 
even_numbers_array.flat[:] = even_numbers[:num_elements]  

print(even_numbers_array)

```

```python
#Task 2 (b)
import numpy as np

second_layer = even_numbers_array[1]  
first_two_rows = second_layer[:2]  

row_means = np.mean(first_two_rows, axis=1)

print(row_means)
```

```python
#Task 2 (c)
import numpy as np
last_three_cols = even_numbers_array[:, :, -3:]
total_sum = np.sum(last_three_cols)
print(total_sum)
```

```python
#Task (d)
import numpy as np
#print(even_numbers_array)
split_arrays = np.split(even_numbers_array, 4, axis=0)
print(split_arrays)

##### Re- explain 
```
