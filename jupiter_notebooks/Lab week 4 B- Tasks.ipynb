# Check for missing values in both Train and Test

```python
import pandas as pd

# Assuming you have loaded the data into a DataFrame named df
missing_values_train = train_df.isnull()
missing_values_test = test_df.isnull()

print("Number of missing values in each column:") #wht are we printing number of missing values in each column 
print(missing_values_train)
print(missing_values_test)
```

# How Much memory is used by Train and Test DataFram

```python
import pandas as pd

# Assuming you have loaded the training and testing data into DataFrames named train_df and test_df
train_df_memory = train_df.memory_usage().sum()
test_df_memory = test_df.memory_usage().sum()

print("Memory usage of the training data:", train_df_memory, "bytes")
print("Memory usage of the testing data:", test_df_memory, "bytes")
```

# Number of null values in each features in both files

```python
import pandas as pd

# Assuming you have loaded the training and testing data into DataFrames named train_df and test_df
train_df_nulls = train_df.isnull().sum()
test_df_nulls = test_df.isnull().sum()

print("Number of null values in the training data:")
print(train_df_nulls)

print("\nNumber of null values in the testing data:")
print(test_df_nulls)
```

# Find the total number of missing values from the both datafram

```python
import pandas as pd

# Assuming you have loaded the training and testing data into DataFrames named train_df and test_df
train_df_missing = train_df.isnull().sum().sum()
test_df_missing = test_df.isnull().sum().sum()

total_missing = train_df_missing + test_df_missing

print("Total number of missing values in the training data:", train_df_missing)
print("Total number of missing values in the testing data:", test_df_missing)
print("Total number of missing values in both datasets:", total_missing)
```
