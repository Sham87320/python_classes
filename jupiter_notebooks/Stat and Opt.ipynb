```python
import numpy as np
import pandas as pd
# for the regression model and lowess smoothing
import statsmodels.api as sm
1
# for stats tests
import scipy.stats as st
# for regression metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
# to find influential data points
from statsmodels.stats.outliers_influence import OLSInfluence
# for diagnostic tests
import statsmodels.stats.diagnostic as di
import statsmodels.stats.stattools as stt
# for general plotting
import matplotlib.pyplot as plt
import plotly.graph_objects as go
# for the linear regression model and splitting data
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#load the data set into Pandas, keeping only the TV and Sales columns
sales_df = pd.read_csv(
"Advertising.csv",
usecols=["TV","sales"]
)
```

```python
#split the data set randomly into train and test partitions, using an 80/20 split.
sales_train, sales_test = train_test_split(
sales_df,
test_size=0.2,
random_state = 42
)
```

```python
# Calculating the slope of a regression line in python B1(Beta 1)

# calculate the deviations of x values from the mean of x
x_deviations = sales_train.TV - sales_train.TV.mean()
# calculate the deviations of y values from the mean of y
y_deviations = sales_train.sales - sales_train.sales.mean()
# calculate Beta1
Beta1 = (x_deviations * y_deviations).sum() / (x_deviations**2).sum()
Beta1
```

```python
#Calculating the intercept B0 (Beta 0)

Beta0 = sales_train.sales.mean() - Beta1*sales_train.TV.mean()
Beta0
```

```python
# The slope is approximately equal to 0.047 and the intercept is approximately equal to 7.12

#Interpreting the slope and intercept
#Sales can be expected to increase by approximately 0.047 when the expenditure on TV advertising goes up by 1 (so there will be no intercept)
#Sales can be expected to decrease by approximately 0.047 when the expenditure on TV advertising goes down by 1.

# Y= B0 + B1X
# So Y = b0 when the tv advertising expenses are zero (The intercept is B0 when X is 0)

```

```python
#Plotting the regression line 
#The regplot() function from the Seaborn library can be used to produce a scatter plot of the x variable against the y variable. It will also plot the regression line (the line of best fit) and a 95%
#confidence band.
import seaborn as sns
sns.set(rc={'figure.figsize':(4,4)})
sns.regplot(x="TV", y="sales", data=sales_train)
```
